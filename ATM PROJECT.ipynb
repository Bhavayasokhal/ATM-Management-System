```python
class ATM:
    "WELCOME TO ATM"
    def __init__(self, balance = 0):
        self.balance = balance
    

    def deposit(self, enter_amount):
        self.balance = self.balance + enter_amount
        print("$",enter_amount,"deposited")
        print("Total Balance = $", self.balance)

    def withdraw(self, enter_amount):
        if enter_amount > self.balance:
            print("INSUFFICIENT BALANCE")
        else:
            self.balance = self.balance - enter_amount
            print("$",enter_amount, "withdrawn")
            print("Remaining Balance = $", self.balance)

    def check_balance(self):
        print("Current balance = $",self.balance)

def atm_session():
    atm = ATM()
    print("*****ATM*****")
    print("d - Deposit")
    print("w - Withdraw")
    print("c - Check Balance")
    print("e - Exit")

    choice = input("Enter your Choice").lower()
    if choice.lower() == "d":
        amount = int(input("Enter the amount to deposit = $"))
        atm.deposit(amount)
    elif choice.lower() == "w":
        amount = int(input("Enter the amount to withdraw = $"))
        atm.withdraw(amount)
    elif choice.lower() == "c":
        atm.check_balance()
    elif choice.lower() == "e":
        print("Thankyou for using ATM")
        return
    else:
        print("Invalid Choice, Try Again")
atm_session()  
```
